```python
from sklearn.datasets import fetch_olivetti_faces  # 머신러닝 lib
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') # 한글 사용
faces = fetch_olivetti_faces()
f, ax = plt.subplots(1,3)
```

```python
ax[0].imshow(faces.images[6],cmap=plt.cm.bone)
ax[0].grid(False)
ax[0].set_xticks([])
ax[0].set_yticks([])
ax[0].set_title('image 첫번째')
```

```python
ax[1].imshow(faces.images[7],cmap=plt.cm.bone)
ax[1].grid(False)
ax[1].set_xticks([])
ax[1].set_yticks([])
ax[1].set_title('image 두번째')
```

```python
new_faces= 0.4 * faces.images[6] + 0.6 * faces.images[7]
ax[2].imshow(new_faces,cmap=plt.cm.bone)
ax[2].grid(False)
ax[2].set_xticks([])
ax[2].set_yticks([])
ax[2].set_title('image 세번째')
plt.show()
f.set_size_inches(10,10)
f
```
